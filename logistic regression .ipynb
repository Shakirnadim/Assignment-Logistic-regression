```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
df=pd.read_csv('bank-full (2).csv',sep=";")
df
```

# EDA

```python
df.isna().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
df1=pd.get_dummies(df,columns=['job','marital','education','contact','poutcome','month'])
df1
```

```python
df1.describe()
```

```python
df1.shape
```

```python
df1['default'] = np.where(df1['default'].str.contains("yes"), 1, 0)
df1['housing'] = np.where(df1['housing'].str.contains("yes"), 1, 0)
df1['loan'] = np.where(df1['loan'].str.contains("yes"), 1, 0)
df1['y'] = np.where(df1['y'].str.contains("yes"), 1, 0)
df1
```

```python
df1.info()
```

# model building

```python
x= pd.concat([df1.iloc[:,0:10],df1.iloc[:,11:]],axis=1)
y=df1.iloc[:,10]
```

```python
x
```

```python
y
```

```python
classifier=LogisticRegression()
classifier.fit(x,y)
```

```python
y_pred=classifier.predict(x)
y_pred
```

```python
y_pred_df=pd.DataFrame({'actual_y':y,'y_pred_prob':y_pred})
y_pred_df
```

# Testing Model Accuracy

```python
confusion_matrix=confusion_matrix(y,y_pred)
confusion_matrix
```

```python
# calculat model accuracy
(39150+1168)/(39150+772+4121+1168)
```

```python
#accuracy=0.8917, which is greater then 0.5: thus [:,1] Threshold value>0.5=0
classifier.predict_proba(x)[:,1]
```

```python
fpr,tpr,thresholds=roc_curve(y,classifier.predict_proba(x)[:,1])
plt.plot(fpr,tpr,color='red')
auc=roc_auc_score(y,y_pred)

plt.plot(fpr,tpr,color='red',label='logit model(area  = %0.2f)'%auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.show()

print('auc accuracy:',auc)
```
